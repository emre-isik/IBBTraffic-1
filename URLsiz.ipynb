```python
import math
from sklearn.metrics.pairwise import haversine_distances

distance_matrix = haversine_distances(
    iskeleler[['lat', 'lon']],
    data[['LATITUDE', 'LONGITUDE']]
)
```

```python
distance_matrix = distance_matrix * 6371  #multiply by Earth radius to get kilometers
```

```python
#Her satır bir iskele. Her sütun, n inci satırdaki iskele koordinatları ile yoğunluğu verilen koordinatların uzaklığı.
print(distance_matrix)
print(distance_matrix.shape)
```

```python
#İlk iskelenin yoğunluklara uzaklığı. Distance matrix'i anlamak adına alttaki kodları yazdım.
print(distance_matrix[0,:])
print(distance_matrix[0].shape)
print(len(distance_matrix))
print(len(distance_matrix[0]))
```

```python
#1km'den küçük olan uzaklıkları(daire içinde olanları) burada topluyorum ve result denen bir array'e koyuyorum.
mLessThan1 = distance_matrix < 1
print(mLessThan1[46,:])
print(mLessThan1.shape)
# Sum elements <1 across rows 
result = np.sum(distance_matrix*mLessThan1, axis=1)
```

```python
#Boyutu doğru mu diye kontrol ediyorum ve topladığım değerlere bakabilirim.
print(result.shape)
print(result)
#8km yarıçap denediğim zaman altta result'ta görüldüğü gibi sadece 1 iskeledeki yoğunluk 0'dan farklı oluyor.
#En sondaki 1.41606.. olan benim denemek için oluşturduğum Deneme iskelesi yoğunluğu.
```

```python
# Her bir iskeledeki en küçük trafik yoğunluk verisini minvar'ın içine koyuyorum
minvar = [999999] * 48

for i in range(0,len(distance_matrix)):
    for j in range(0,len(distance_matrix[i,:])):
        if(distance_matrix[i,j] < distance_matrix[i,j-1]):
            minvar[i] = distance_matrix[i,j]

```

```python
#Her iskeledeki verilen en küçük yoğunluk verisini beyaz circle ile çiziyorum. Böylece iskele circle'larının (yeşil circle)   
#içinde olup olmayacağını anlayabileceğim. En küçük yoğunluk verilerini (minvar) değerlerini yazdırıyorum.
for i in range(0,len(iskeleler)):
   folium.Circle(
      location=[iskeleler.iloc[i]['lat'], iskeleler.iloc[i]['lon']],
      radius=minvar[i],  #meter
      color='#fff',
      fill=True,
      fill_color='#fff',
      fill_opacity=0.1
   ).add_to(harita)

print(minvar)
# Show the map again
harita
```
